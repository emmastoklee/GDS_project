## Exploratory analysis of distance graph analysis

```python
import pandas as pd
```

```python
driving = pd.read_csv('../app/distances/all_driving_distances.csv')
walking = pd.read_csv('../app/distances/all_walking_distances.csv')
biking = pd.read_csv('../app/distances/all_biking_distances.csv')
```

### Biking analysis

```python
# group data by neighbourhood and amenity and calculate the avg. travel time per group
grouped_data = biking.groupby(['neighborhood', 'amenity']).travel_time.mean().reset_index()
# Sorting the grouped data by average travel time in descending order and selecting the top 10
grouped_data.sort_values(by='travel_time', ascending=False).head(10)
```

```python
# Grouping the data by amenity and calculating the mean travel time
amenity_avg_travel_time = biking.groupby('amenity').travel_time.mean().reset_index()
# Sorting the data by average travel time in descending order and selecting the top 3
amenity_avg_travel_time.sort_values(by='travel_time', ascending=False).head(10)
```
