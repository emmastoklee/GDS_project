```python
import geopandas as gpd # for geospatial data handling
import pandas as pd
import osmnx as ox # for handling data from OpenStreetMap (osm) with the help of networkX (nx)
import contextily as cx # for plotting
import matplotlib.pyplot as plt # for plotting
from pyproj import CRS # for more advanced CRS modifications and transformations
from shapely.geometry import Polygon
import numpy as np
```

```python
# Define the city name or bounding box
city = 'Montreal, Canada'
```

```python
pois = ox.geometries_from_place(
    city, tags={"amenity": True}
)
```

```python
len(pois)
```

```python
len(pois['amenity'].value_counts())
```

```python
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Display the full list of value counts for the 'amenity' column
print(pois['amenity'].value_counts())
```

```python
pois.head()
```

```python
# remove any poi amenity types where there are less than 10 of that type
clean_pois = pois.groupby('amenity').filter(lambda x: len(x) >= 10)
```

```python
clean_pois.head()
```

```python
len(clean_pois)
```

```python
len(clean_pois.columns)
```

```python
print(clean_pois.columns.tolist())
```

```python
# keep only relevant columns
clean_pois = clean_pois[['direction', 'geometry', 'amenity', 'name']]
```

```python
# delete amenities we are not interested in 
clean_pois = clean_pois[~clean_pois['amenity'].isin(['parking','bench', 'bicycle_parking', 'waste_basket', 'parking_space', 'charging_station', 'drinking_water', 'taxi', 'loading_dock', 'parking_entrance', 'vending_machine', 'telephone', 'shelter', 'toilets', 'fuel', 'bicycle_rental', 'fountain', 'public_bookcase', 'lounger', 'bicycle_repair_station', 'car_wash', 'car_rental', 'car_sharing', 'table', 'driving_school', 'dojo', 'waste_disposal', 'bus_station', 'smoking_area', 'money_transfer', 'motorcycle_parking'])]
```

```python
# check remaining amenities
clean_pois['amenity'].value_counts()
```

```python
# put amenities is new categories so we have a lower number of categories
# Define the conditions and corresponding categories
conditions = [
    clean_pois['amenity'].isin(['restaurant', 'cafe', 'fast_food', 'ice_cream']),
    clean_pois['amenity'] == 'school',
    clean_pois['amenity'].isin(['university', 'college']),
    clean_pois['amenity'].isin(['bar', 'pub', 'nightclub']),
    clean_pois['amenity'].isin(['pharmacy', 'dentist', 'clinic', 'hospital']),
    clean_pois['amenity'].isin(['kindergarten', 'childcare']),
    clean_pois['amenity'].isin(['theatre', 'cinema']),
    clean_pois['amenity'] == 'place_of_worship',
    clean_pois['amenity'].isin(['bank', 'atm']),
    clean_pois['amenity'].isin(['community_centre', 'library', 'arts_centre', 'events_venue']),
    clean_pois['amenity'].isin(['social_facility', 'fire_station', 'police', 'recycling', 'social_center', 'public_building']),
    clean_pois['amenity'].isin(['post_box', 'post_office'])
]

categories = [
    'Restaurant/cafe',
    'School',
    'Further Education',
    'Nightlife',
    'Health care',
    'Child support',
    'Entertainment',
    'Religious place',
    'Bank/atm',
    'Culture',
    'Public services',
    'Post'
]

# Create a new column 'categories' based on the conditions and categories
clean_pois['category'] = np.select(conditions, categories, default=None)


```

```python
clean_pois.head()
```

```python
# save clean version of geodataframe
clean_pois.to_file('clean_pois_montreal.geojson', driver='GeoJSON')

```
